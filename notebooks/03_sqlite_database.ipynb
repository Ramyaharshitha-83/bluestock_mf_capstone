```python
import sqlite3
from pathlib import Path

repo_root = Path.cwd()
if repo_root.name == "notebooks":
    repo_root = repo_root.parent

db_path = repo_root / "data" / "db"
db_path.mkdir(parents=True, exist_ok=True)

conn = sqlite3.connect(str(db_path / "bluestock_mf.db"))

with open(str(repo_root / "sql" / "schema.sql"), "r") as f:
    conn.executescript(f.read())

conn.commit()
conn.close()

print("Database created successfully")
```

```python
import pandas as pd
import sqlite3

conn = sqlite3.connect("/workspaces/bluestock_mf_capstone/data/db/bluestock_mf.db")

nav = pd.read_csv("/workspaces/bluestock_mf_capstone/data/processed/clean_nav.csv")
tx = pd.read_csv("/workspaces/bluestock_mf_capstone/data/processed/clean_transactions.csv")
perf = pd.read_csv("/workspaces/bluestock_mf_capstone/data/processed/clean_performance.csv")

nav.to_sql("fact_nav", conn, if_exists="replace", index=False)
tx.to_sql("fact_transactions", conn, if_exists="replace", index=False)
perf.to_sql("fact_performance", conn, if_exists="replace", index=False)

conn.close()

print("Data loaded successfully")
```
